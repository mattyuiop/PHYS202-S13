```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
theta = linspace(-pi,pi, 20*pi)
```

```python
y = (cos(2*theta)*cos(theta/2))**2
plot(theta,y, 'r')
plot(theta,y, '*b')
xtic = [-pi, -pi/2, -pi/4,0,pi/4,pi/2,pi]
xticlab = ["$\pi$", r"-$ \frac{\pi}{2}$",r"-$\frac{\pi}{4}$","0",r"$\frac{\pi}{4}$",r"$\frac{\pi}{2}$","${\pi}$"]
xax = r"$\theta$" + ' (rad)'
xlabel(xax)
xticks(xtic,xticlab)
ylabel('intensity')
show()
#print xticlab
```

```python
x = linspace(-5,5,)
by = (x+1)/(x-1)
ry = exp(-x)
plot(x,by, 'b', label = '(1+t)/(t-1)')
plot(x,ry, ':r', label = 'exp(-t)')
xlabel('time (seconds)')
ylabel('f(t), g(t)')
xlim([-4.5,4.5])
ylim([-10,10])
legend(loc = 3)
show()
```

```python
def sumChrg(x,y,a,b,Xo,Yo,cd,axis):
    dx = .001
    tot = 0*x
    for i in range(int(a*1000),int((b+dx)*1000),1):
        p = i/1000
        tot += pointField(x,y,cd*dx,p,0)[axis]
    return tot[Xo,Yo]
sumChrg(x,y,1,2,3,3,1e-6,0)
```

```python
sumChrg(x,y,1,2,3,3,1e-6,1)
```

```python
def sumChangChrg(x,y,a,b,Xo,Yo,cd,axis):
    dx = .001
    tot = 0*x
    for i in range(int(a*1000),int((b+dx)*1000),1):
        p = i/1000
        tot += pointField(x,y,cd*dx*p**2*3**-2,p,0)[axis]
    return tot[Xo,Yo]
```

```python
a = sumChangChrg(x,y,1,3,5,3,2e-6,0)
b = sumChangChrg(x,y,1,3,5,3,2e-6,1)
```
