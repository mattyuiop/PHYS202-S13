$ \sum F_{tang} = -mg sin \theta  = ma_{tang}$


$\frac{d^2s}{dt^2} = l \frac{d^2s}{dt^2} $

$\frac{d\theta}{dt} = \omega $

$\frac{d\omega}{dt} =\frac{g}{l} sin \theta$
 


```python
g=9.8 #m/s^2
l=0.5 #m
thO = 20*pi/180 #20 deg
tO,tf,dt=0.,10.,.01
t = arange(tO,tf+dt,dt)
theta_app = lambda t: thO*cos(t*(g/l)**0.5)
#numerical:
def deriv(theta_current,t): #returns derivative array
    g = 9.8
    l = 0.5
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/l)*sin(theta)
    return array([dtheta,domega])

from scipy.integrate import odeint

t = arange(tO,tf+dt,dt)
omegaO = 0.
theta_init = array([thO,omegaO])

theta = odeint(deriv,theta_init,t)
    
               
plot(t,theta_app(t),'b-', t,theta[:,0], 'r--')
#(t,theta_app(t))
show()
```

```python
thO = 50.*pi/180
theta_init = array([thO,omegaO])
theta = odeint(deriv,theta_init,t)
plot(t,theta_app(t),'b-', t,theta[:,0], 'r--')
plot()
```

```python
thOo,thOt,thOtr,thOfr,thOfv = 0.0, pi/(4),pi/(2),pi*3/(4),pi
theta_init = array([thOo,omegaO])
theta = odeint(deriv,theta_init,t)

theta_initt = array([thOt,omegaO])
thetat = odeint(deriv,theta_initt,t)

theta_inittr = array([thOtr,omegaO])
thetatr = odeint(deriv,theta_inittr,t)

theta_initfr = array([thOfr,omegaO])
thetafr = odeint(deriv,theta_initfr,t)

theta_initfv = array([thOfv,omegaO])
thetafv = odeint(deriv,theta_initfv,t)

plot(t,theta_app(t),'b:', t,theta[:,0], 'r--',t,thetat[:,0], 'm--',t,thetatr[:,0], ':',t,thetafr[:,0],
 'c--',t,thetafv[:,0], '--')
plot()
```
