##Prime numbers

```python
is_prime = np.ones((100,), dtype=bool)
is_prime[0] = 0
is_prime[1] = 0

N_max = int(np.sqrt(len(is_prime)))
for j in range(2, N_max):
    is_prime[2*j::j] = False

print np.nonzero(is_prime)
```

# the for loop works by 'falsing' every multiple of the numbers from 2 to 10
# the loop only needs to work to 10 or sqrtN since any composite number must have a factor between 
# 2 and  sqrt(N) to be within range.  higher multiples of primes like 23*3 is falsed when j = 3 and the loop counts to the end in steps of j 

```python
def eratosthenes_sieve(N):
    is_prime = np.ones((N,),dtype = bool)
    is_prime[0] = 0
    is_prime[1] = 0
    N_max = int(np.sqrt(len(is_prime))+1)
    for j in range(2,N_max):
        is_prime[2*j::j] = False
    return np.nonzero(is_prime)
```

```python
print eratosthenes_sieve(200)
```
