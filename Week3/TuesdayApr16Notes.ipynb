```python
t = arange(0.0, 1.0 + .01, 0.01)
c = cos(2*2*pi*t)
s = sin(4*pi*t)
```

```python
print pi
```

```python
plot(t,c, color = 'turquoise',linestyle='--', linewidth = 2,label="$cos({\pi} t)$")
plot(t,s, color = 'magenta')
xlabel('time [s]')
ylabel('Voltage [mV]')
#Add title for convenience
title('A simple plot', fontsize = 50)
grid(True)
legend(loc = 'best')
```
