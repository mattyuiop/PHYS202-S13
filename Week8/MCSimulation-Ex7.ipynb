##Weibull distribution

"$ f(x)$ = {$ \alpha $ $ \beta ^ {-\alpha} x ^{\alpha - 1} e^{-(x/ \beta)^\alpha }$ if $x>0 $" 
$ f(x) = 0 $  otherwise

$\int \limits^x_{-\infty}  f(x)$ =  $ F(x) = 1 - e^{ -(x/ \beta ) ^ {-\alpha}} $

$F^{-1}(x) = -\beta [ln( 1 - x )]^{1/\alpha} $

```python
# analytic solution:
a=b=alpha = beta= 1

f = lambda x:a*b**(-a)*x**(a-1)*e**(-1* (x/b)**a)
F = lambda x: 1-e**(- (x/b)**a)
Finv = lambda x: -b*(log(1-x)**(1/a))
x = arange(0,1000,.1)
```

```python
def gimmeRnd(n):
    """My random numb Generator manipulates time on scale less than miliseconds and carries 
    a random b the further differentiate between miliseconds and the offset calcualted one line after
    """
    import time
    allofEm = []
    b=0
    for i in range(n):
        a = time.time()
        millis = time.time() * 1000000.0 % 100.0 # takes a small fraction of a second
        milloffset = time.time()*10.0**9 % 100.0 + b
        new = millis/milloffset
        if abs(new) < 1:
            allofEm = append(allofEm,new)
            b = new**(time.time()*999 % 1000.0)
        else:
            new = new - floor(new)
            allofEm = append(allofEm,new)
            b = new**(time.time()*999 % 1000.0)
            #print 'shit',
        
    return allofEm
    
#h = gimmeRnd(12)  
#print h
#time.time()
```
