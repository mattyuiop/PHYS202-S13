```python
def distribute1D(x,prob,N):
    """ takes any numerical distribution prob, on the interval defined by the array x and returns an aaray of N sampled 
    values there are statistically the same as the input data."""
    numCDFs = cumsum(prob) / sum(prob)

    z = numpy.random.random_sample(N)
    
    a = numCDFs.searchsorted(z) #gives locations in numCDF corresponding to random array
    sampledXs = x[a]
    
    return sampledXs
```

```python
sampledXs = distribute1D(x,fs(x),10**6)

hist(sampledXs,50,normed = True, label = 'Numrical Dist')
plot(x,CDF(x), label = 'analytic CDF', lw=3, color= 'red')
plot(x,fs(x), label = 'f(x)', lw=3, color= 'm')
xlim(0,pi)
ylim(0,1)
legend(loc = 0)
show()
```

```python
d = 15*10**-6
a = 10**-5
A = 2.0
lamd = 632.8 *10**-9
L = 1.
x = arange(-.2,.2,10**-6)
Iss = lambda x: A**2*(sinc(a*x/(lamd*L)))**2.
Ids = lambda x: A**2*(cos(2*pi*d*x/(lamd*L)))**2 
Itot = lambda x: Iss(x) * Ids(x)
#Inew = A * (cos())
```
