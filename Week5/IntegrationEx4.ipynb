```python
L = 1.0
PsiSqGrnd = lambda x: (2/L) * sin(1*x*pi/L)**2

PsiSqFirst = lambda x: (2/L) * sin(2*x*pi/L)**2

FoX = lambda x: e**(-x**2)
def intEofX(fxn, x0, xf, step):
    """Integral approx with Left hand Rehiman sum"""
    maxcount = (xf - x0)/step
    x = arange(x0, xf+step,step)
    ingral = zeros(maxcount +1)
    for i in range(int(maxcount+1)):
        if i ==0:
            ingral[i] = fxn(i)*step
        else:
            ingral[i] = fxn(i)*step + ingral[i-1]
    plot(x,ingral)
    return ingral[maxcount], size(ingral),maxcount
    
```
